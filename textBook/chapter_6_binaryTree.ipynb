```python
def create_binaryTree(tree, data):
    for i in range(len(data)):
        level = 0 
        if i == 0:
            tree[level] = data[i]   # 第 0 索引資料放在第 0 層
        else:
            while tree[level]:      # 當 while 迴圈結束表示找到存放數據的節點(索引)的位置
                if data[i] > tree[level]:         # 如果資料大於節點索引, 往右搜尋
                    level = level*2+2
                else:                             # 否則往左搜尋
                    level = level*2+1
        tree[level] = data[i]                     # 找到數據應存放的節點索引
        print(i, tree)                            # 建立二元樹的過程

binary_tree = [0]*8
data = [10, 21, 5, 9, 13, 28]
create_binaryTree(binary_tree, data)
for i in range(len(binary_tree)):
    print("二元樹陣列btree[%d] = %d"%(i, binary_tree[i]))
```

ch 6-6-2 Using linked-list to create binary tree
inorder 運算子在中間 (print)
preorder 運算子在前面 
postorder 運算子在後面

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None

    def insert(self, data):
        if self.data:                           # 如果 root 存在
            if data < self.data:                # 如果小於目前 node 值
                if self.left:                   # 如果目前 node 左節點存在
                    self.left.insert(data)      # 遞迴往下一層
                else:
                    self.left = Node(data)      # 不存在則將值丟給左節點
            else:
                if self.right:                  # 如果 right node 存在
                    self.right.insert(data)     # 遞迴往下一層
                else:
                    self.right = Node(data)     # 不存在則將值丟給右節點
        else:                                   # 如果根結點不存在
            self.data = data                    # 建立根結點

    def inorder(self):
        if self.left:                           # 左子樹節點存在
            self.left.inorder()                 # 遞迴往下一層走
        print(self.data)                        # 左子樹已不存在, 處理此節點(print)
        if self.right:                          # 如果右子樹存在
            self.right.inorder()                # 遞迴往下一層走
    def preorder(self):
        print(self.data)
        if self.left:
            self.left.preorder()
        if self.right:
            self.right.preorder()
    def postorder(self):
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        print(self.data)


tree = Node()
datas = [10, 21, 5, 9, 13, 28]
for d in datas:
    tree.insert(d)
print('中序')
tree.inorder()
print('-'*50)
print('前序')
tree.preorder()
print('-'*50)
print('後序')
tree.postorder()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left:
                    self.left.insert(data)
                else:
                    self.left = Node(data)
            else:
                if self.right:
                    self.right.insert(data)
                else:
                    self.right = Node(data)
        else:
            self.data = data
    def search(self, val):
        global step
        if val < self.data:
            if self.left:
                step+=1
                return self.left.search(val)
            else:
                return str(val)+"不存在"
        elif val > self.data:
            if self.right:
                step+=1
                return self.right.search(val)
            else:
                return str(val)+"不存在"
        else:
            return str(val)+'找到了共'+str(step)+"步"

step = 0
tree = Node()
datas = [10, 21, 5, 9, 13, 28]
for d in datas:
    tree.insert(d)

n = eval(input('請輸入愈搜尋的資料'))
tree.search(n)
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left:
                    self.left.insert(data)
                else:
                    self.left = Node(data)
            else:
                if self.right:
                    self.right.insert(data)
                else:
                    self.right = Node(data)

        else:
            self.data = data

    def inorder(self):
        if self.left:
            self.left.inorder()
        print(self.data)
        if self.right:
            self.right.inorder()
class Delete_Node():
    def deleteNode(self, root ,key):
        if root is None:                                      # 二元樹不存在返回
            return None              
        if key < root.data:                                   # 刪除值小於 root 值則往左
            root.left = self.deleteNode(root.left, key)
            return root
        if key > root.data:                                   # 刪除值大於 root 則往右
            root.right = self.deleteNode(root.right, key)
            return root
        if root.left is None:                                 # 左邊節點不存在
            new_root = root.right                             # 改成去找右邊子樹
            return new_root
        if root.right is None:                                # 右邊節點不存在
            new_root = root.left                              # 改成去找左邊子樹
            return new_root

        succ = self.max_node(root.left)                       # 尋找左子樹最大節點
        tmp = Node(succ.data)                                 # 用最大值建立節點
        tmp.left = self.left_node(root.left)                  # 串接原刪除的節點的左子樹
        tmp.right = root.right                                # 串接原刪除的右子樹
        return tmp

    def left_node(self, node):
        if node.right is None:                                # 右子節點不存在
            new_root = node.left                              # 使用左子節點
            return new_root
        node.right = self.left_node(node.right)               # 進入下一層
        return node

    def max_node(self, node):
        while node.right:                                     # 如果是否則 node 是最大值節點
            node = node.right
        return node

tree = Node()
datas = [10, 5, 21, 9, 13, 28, 3, 4, 1, 17, 32]
for d in datas:
    tree.insert(d)
tree.inorder()
del_data = 5
print('刪除 %d 資料後'%del_data)
delete_obj = Delete_Node()
result = delete_obj.deleteNode(tree, del_data)
result.inorder()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left:
                    self.left.insert(data)
                else:
                    self.left = Node(data)
            else:
                if self.right:
                    self.right.insert(data)
                else:
                    self.right = Node(data)
        else:
            self.data = data
    def postorder(self):
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        print(self.data)

class Delete_Node:
    def deleteNode(self, root, Key):
        if root.data == None:
            return None
        if Key < root.data:
            root.left = self.deleteNode(root.left,Key)
            return root
        if Key > root.data:
            root.right = self.deleteNode(root.right,Key)
            return root
        if root.left is None:
            new_node = root.right
            return new_node
        if root.right is None:
            new_node = root.left
            return new_node
        succ = self.min_node(root.right)
        tmp = Node(succ.data)
        tmp.left = root.left
        tmp.right = self.right_node(root.right)
        return tmp
    def right_node(self, node):
        if node.left is None:
            new_root = node.right
            return new_root
        node.left = self.right_node(node.left)
        return node
    def min_node(self, node):
        while node.left:
            node = node.left
        return node


tree = Node()
datas = [10, 5, 21, 9, 13, 28, 3, 4, 1, 17, 32]
for d in datas:
    tree.insert(d)
tree.postorder()
del_data = 10
print('刪除 %d 資料後'%del_data)
delete_obj = Delete_Node()
result = delete_obj.deleteNode(tree, del_data)
result.postorder()
```
