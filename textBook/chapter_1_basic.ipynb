factorial (階層)

```python
def factorial(n):
    if n == 1 :
        return 1
    else : 
        return n*factorial(n-1)
```

```python
print(factorial(30))
```

假設每秒可以處理10兆個數列，計算最差要花多少年才可以得到從小排到大的數列

```python
year = 365
hours = 24
minutes = 60
sec = 60

fac = factorial(30)
print(int(fac/(year*hours*minutes*sec)),'years')
```

itertools 中 permutation( ) 可以枚舉所有可能排列

```python
import itertools

x = [i for i in range(1,4)]
perm = itertools.permutations(x)
for i in perm:
    print(i)
```

繪製時間複雜度關係圖 O(n), O(logn), O(1)

```python
import matplotlib.pyplot as plt
import numpy as np

xpt = np.linspace(1, 10, 10) # 建立含 10 個元素的陣列
ypt1 = xpt/xpt  # O(1)
ypt2 = np.log(xpt)  # O(logn)
ypt3 = xpt
ypt4 = xpt**2
ypt5 = [factorial(i) for i in xpt]

plt.plot(xpt, ypt1, '-o', label="O(1)")
plt.plot(xpt, ypt2, '-o', label="O(logn)")
plt.plot(xpt, ypt3, '-o', label="O(n)")
#plt.plot(xpt, ypt4, '-o', label="O(n**2)")
#plt.plot(xpt, ypt5, '-o', label="O(n!)")
plt.legend(loc="best")

plt.show()
```

繪製 O(1), O(logn), O(n), O(nlogn), O(n**2)

```python
import matplotlib.pyplot as plt
import numpy as np

xpt = np.linspace(1, 10, 10) # 建立含 10 個元素的陣列
ypt1 = xpt/xpt  # O(1)
ypt2 = np.log(xpt)  # O(logn)
ypt3 = xpt
ypt4 = xpt*ypt2
ypt5 = xpt**2

plt.plot(xpt, ypt1, '-o', label="O(1)")
plt.plot(xpt, ypt2, '-o', label="O(logn)")
plt.plot(xpt, ypt3, '-o', label="O(n)")
plt.plot(xpt, ypt4, '-o', label="O(nlogn)")
plt.plot(xpt, ypt5, '-o', label="O(n**2)")
plt.legend(loc="best")

plt.show()
```

### 1-6 習題

20 筆資料幾種排列

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n*factorial(n-1)


print(factorial(20))
```

延續上面，產生一個組合要 0.0000000001，產生所有組合需要多少時間

```python
total_step = factorial(20)
time = 0.0000000001
print(int(total_step*time),'sec')
```

列出 a, b, c, d, e, f 所有排列

```python
import itertools
arr = ['a','b','c','d','e','f']
perms = itertools.permutations(arr)
n=0
for perm in perms:
    n+=1
    print(perm)
print(f'總共有 {n} 種排列方式')
```
